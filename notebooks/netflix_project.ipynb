Dataset Overview

Source: Netflix Movies and TV Shows Dataset on Kaggle
Description:
This dataset contains metadata about Netflix’s catalog of movies and TV shows. It is ideal for exploring trends in content, genres, and ratings.
Key Features:

    show_id: Unique identifier for each movie or TV show.
    type: Whether the content is a "Movie" or a "TV Show".
    title: The title of the content.
    director: Name of the director (if available).
    cast: Key cast members.
    country: Country where the content was produced.
    date_added: When the content was added to Netflix.
    release_year: The release year of the movie or show.
    rating: Content rating (e.g., TV-MA, PG-13).
    duration: Duration of movies (in minutes) or TV shows (number of seasons).
    listed_in: Categories/genres of the content.
    description: A short summary of the content.

```python
# Fill or drop missing values:

# - Replace missing values in 'rating' and 'director' with 'Unknown'

netflix_data.fillna({'rating': 'Unknown'}, inplace=True)
netflix_data.fillna({'director': 'Unknown'}, inplace=True)

netflix_data
```

```python
# - Drop rows where 'date_added' is missing
netflix_data = netflix_data.dropna(subset=['date_added'])
netflix_data
```

```python
# Remaining columns with missing values are filled with placeholders

netflix_data.fillna({'country': 'Unknown'}, inplace=True)
netflix_data.fillna({'cast': 'Unknown'}, inplace=True)
netflix_data
```

```python
# Convert 'date_added' to datetime

netflix_data['date_added']= pd.to_datetime(netflix_data['date_added'], errors='coerce')

#Extract the year and month for trend analysis

netflix_data['release_year'] = netflix_data['date_added'].dt.year
netflix_data['release_month'] = netflix_data['date_added'].dt.month_name
```

```python
# Split 'duration' into numeric values and clean the column
def extract_duration(duration):
    if isinstance(duration, str):
        # Extract the numeric part (e.g., "90 min" -> 90, "1 Season" -> 1)
        return int(duration.split()[0])
    return 0

# Create a new column for numeric duration
netflix_data['duration_num'] = netflix_data['duration'].apply(extract_duration)
netflix_data
# here ignoring the fact that the presumed analysis for this might be wrong, but the that is taken care of because the difference arrises with the movies and Tv shows which already has column for this so no issues with this 
```

```python
# Convert 'listed_in' column into a list of genres
netflix_data['genres'] = netflix_data['listed_in'].str.split(', ')
```

## 1. Distribution of Movies vs. TV Shows
### We will calculate the count of Movies and TV Shows in the dataset and visualize the results with a bar chart.

```python
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Count the distribution of 'type'
type_distribution = netflix_data['type'].value_counts().reset_index()
type_distribution.columns = ['type', 'count']

# Plot the distribution as a bar chart
plt.figure(figsize=(8, 6))
sns.barplot(
    data=type_distribution,
    x='type',
    y='count',
    hue='type',  # Assign the same column to hue
    palette="coolwarm",
    dodge=False  # Prevents splitting of bars by hue
)
plt.title('Distribution of Movies vs. TV Shows on Netflix', fontsize=16)
plt.xlabel('Type', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title='Type')  # Keeps the legend if you want to display it
plt.show()
```

### The bar chart shows the count of Movies and TV Shows in the dataset. As expected, Netflix has a significantly higher number of movies compared to TV shows. This aligns with Netflix's focus on a diverse catalog but with more movies historically.

## 2. Growth of Netflix Catalog Over Time
### We'll analyze how Netflix's catalog has grown over time by visualizing the number of titles added each year

```python
# Count the number of titles added each year
titles_per_year = netflix_data['release_year'].value_counts().sort_index()

# Plot the trend of titles added over time
plt.figure(figsize=(10, 6))
sns.lineplot(
    x=titles_per_year.index,
    y=titles_per_year.values, 
    marker="o", 
    color="royalblue"
)
plt.title('Growth of Netflix Catalog Over Time', fontsize=16)
plt.xlabel('Year Added', fontsize=14)
plt.ylabel('Number of Titles', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### A noticeable spike in the number of titles added around recent years, which reflects Netflix's expansion efforts. Before a certain period (likely pre-2015), the growth was relatively slow, as Netflix was transitioning from DVD rentals to a streaming platform.

## 3. Distribution of Content Ratings
### We will explore the distribution of content ratings to understand the variety and age-appropriateness of titles on Netflix. Let's proceed!

```python
# Check for unique values in the 'rating' column
print("Unique values in 'rating':")
print(netflix_data['rating'].unique())

# list of valid ratings based on observations
valid_ratings = [
    'TV-MA', 'TV-14', 'R', 'PG-13', 'PG', 'TV-PG', 
    'TV-Y', 'TV-Y7', 'TV-G', 'NR', 'G', 'UR', 'NC-17'
]

#  Drop rows with invalid or missing ratings
netflix_data_cleaned = netflix_data[netflix_data['rating'].isin(valid_ratings)].copy()

#Verify the updated dataset
print(f"Number of rows before cleaning: {len(netflix_data)}")
print(f"Number of rows after cleaning: {len(netflix_data_cleaned)}")

# Optional: Reset the index for the cleaned dataset
netflix_data_cleaned.reset_index(drop=True, inplace=True)

# Display the cleaned dataset
netflix_data_cleaned.head()

# Count the distribution of content ratings
rating_distribution = netflix_data_cleaned['rating'].value_counts()

# Plot the distribution of ratings as a bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=rating_distribution.index,
            hue= rating_distribution.index,
            y=rating_distribution.values, 
            palette="viridis")
plt.title('Distribution of Content Ratings on Netflix', fontsize=16)
plt.xlabel('Rating', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### TV-MA (Mature Audience) and TV-14 dominate the catalog, indicating a focus on mature and teen-friendly content. A significant number of titles are marked Unknown, likely due to missing metadata.

## 4. Most Common Genres
### Let's examine the most common genres/categories on Netflix by analyzing the listed_in column. This will help identify trends in content preferences


```python
from collections import Counter

# Flatten the list of genres and count occurrences
all_genres = [genre for sublist in netflix_data['genres'].dropna() for genre in sublist]
genre_counts = Counter(all_genres).most_common()

# Convert to a DataFrame for visualization
genre_df = pd.DataFrame(genre_counts, columns=['Genre', 'Count'])

# Plot the most common genres
plt.figure(figsize=(12, 8))
sns.barplot(y=genre_df['Genre'], hue=genre_df['Genre'], x=genre_df['Count'], palette="mako")
plt.title('Most Common Genres on Netflix', fontsize=16)
plt.xlabel('Count', fontsize=14)
plt.ylabel('Genre', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

### Dramas, Comedies, and Documentaries are the leading genres, indicating their broad appeal to Netflix's audience. Categories like International Movies and Children & Family Movies also feature prominently, showcasing Netflix's diverse content offerings.

## 5. Top 10 Country-wise Distribution of Content
### we'll analyze the country-wise distribution of content to determine which regions dominate Netflix's catalog.

```python
# Count the number of titles by country
country_distribution = netflix_data['country'].value_counts().head(10)  # Top 10 countries

# Plot the country-wise distribution
plt.figure(figsize=(12, 6))
sns.barplot(y=country_distribution.index,hue=country_distribution.index, x=country_distribution.values, palette="cubehelix")
plt.title('Top 10 Countries by Number of Titles on Netflix', fontsize=16)
plt.xlabel('Count', fontsize=14)
plt.ylabel('Country', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

### United States leads by a wide margin, reflecting its dominance in content production. Other countries like India, United Kingdom, and Canada contribute significantly, highlighting Netflix's investment in international content.

## 6. Most Prolific Directors
### We'll now identify the most prolific directors and actors featured in Netflix's catalog. This analysis will uncover the creators and stars who shape the platform's offerings.


```python
# Count the frequency of directors (excluding 'Unknown')
director_counts = netflix_data[netflix_data['director'] != 'Unknown']['director'].value_counts().head(10)

# Plot the most prolific directors
plt.figure(figsize=(12, 6))
sns.barplot(y=director_counts.index, hue=director_counts.index, x=director_counts.values, palette="magma")
plt.title('Top 10 Most Prolific Directors on Netflix', fontsize=16)
plt.xlabel('Count', fontsize=14)
plt.ylabel('Director', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

### Directors with extensive catalogs, such as those specializing in documentaries, TV shows, or popular movies, dominate this list. The data may reflect Netflix's collaborations with these creators for multiple projects.

## 7. Most Frequently Featured Actors
### we'll examine the most frequently featured actors on Netflix, which can give us insights into audience preferences and casting trends

```python
# Flatten the list of actors from the 'cast' column
all_actors = [actor for sublist in netflix_data[netflix_data['cast'] != 'Unknown']['cast'].str.split(', ') for actor in sublist if isinstance(sublist, list)]
actor_counts = Counter(all_actors).most_common(10)

# Convert to a DataFrame for visualization
actor_df = pd.DataFrame(actor_counts, columns=['Actor', 'Count'])

# Plot the most frequently featured actors
plt.figure(figsize=(12, 6))
sns.barplot(y=actor_df['Actor'], hue= actor_df['Actor'], x=actor_df['Count'], palette="plasma")
plt.title('Top 10 Most Frequently Featured Actors on Netflix', fontsize=16)
plt.xlabel('Count', fontsize=14)
plt.ylabel('Actor', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

### These actors likely participate in ensemble casts, long-running TV shows, or are frequent collaborators in Netflix projects. The diversity of actors reflects Netflix's global audience and content strategy.

## 8. Duration Trends
### We'll analyze duration trends to identify the most common lengths of movies and the typical number of seasons for TV shows

```python
# Separate Movies and TV Shows for duration analysis
movies = netflix_data[netflix_data['type'] == 'Movie']
tv_shows = netflix_data[netflix_data['type'] == 'TV Show']

# Analyze movie durations
movie_duration_counts = movies['duration_num'].value_counts().sort_index()

# Analyze TV show durations (seasons)
tv_show_seasons_counts = tv_shows['duration_num'].value_counts().sort_index()

# Plot the duration trends for Movies
plt.figure(figsize=(12, 6))
sns.lineplot(x=movie_duration_counts.index, y=movie_duration_counts.values, marker="o", color="coral", label="Movies")
plt.title('Movie Durations on Netflix', fontsize=16)
plt.xlabel('Duration (Minutes)', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(fontsize=12)
plt.show()

# Plot the number of seasons for TV Shows
plt.figure(figsize=(12, 6))
sns.barplot(x=tv_show_seasons_counts.index, hue=tv_show_seasons_counts.index, y=tv_show_seasons_counts.values, palette="viridis")
plt.title('Number of Seasons for TV Shows on Netflix', fontsize=16)
plt.xlabel('Number of Seasons', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Movie Durations: The majority of movies have durations around 90–120 minutes, reflecting standard feature-length films. Some outliers with very short or very long durations may include documentaries or extended cuts.

### TV Show Seasons:  Most TV shows have 1–2 seasons, likely reflecting new shows or limited series. A smaller number of TV shows extend to higher season counts, indicating long-running series.

## 9. Ratings in Movies vs TV Shows
### Compare how ratings (e.g., TV-MA, PG) are distributed between Movies and TV Shows to identify content type preferences.

```python
rating_by_type = netflix_data_cleaned.groupby(['type', 'rating']).size().unstack(fill_value=0)

# Plot stacked bar chart
rating_by_type.T.plot(kind='bar', stacked=True, figsize=(12, 6), color=['#1f77b4', '#ff7f0e'])
plt.title('Ratings in Movies vs TV Shows', fontsize=16)
plt.xlabel('Rating', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Type')
plt.xticks(rotation=45)
plt.show()
```

### The analysis shows that TV-MA is the dominant rating for both Movies and TV Shows, reflecting Netflix's focus on adult-oriented content. Movies have a broader range of ratings, including family-friendly options, while TV Shows primarily target mature audiences with fewer lower-rated titles.

## 10. Relationship Between Genres and Ratings
### We'll investigate the relationship between genres and ratings, analyzing how different types of content are rated. This can provide insights into audience preferences and content guidelines

```python
# Explode the genres to analyze their relationship with ratings
genre_rating_data = netflix_data.explode('genres')[['genres', 'rating']]

# Count the occurrences of each rating within each genre
genre_rating_counts = genre_rating_data.groupby(['genres', 'rating']).size().unstack(fill_value=0)

# Select a few popular genres for a focused visualization
top_genres = genre_rating_data['genres'].value_counts().head(10).index
filtered_genre_ratings = genre_rating_counts.loc[top_genres]

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(filtered_genre_ratings, cmap="YlGnBu", annot=True, fmt="d", cbar=True)
plt.title('Relationship Between Genres and Ratings', fontsize=16)
plt.xlabel('Rating', fontsize=14)
plt.ylabel('Genre', fontsize=14)
plt.xticks(fontsize=12, rotation=45)
plt.yticks(fontsize=12)
plt.show()
```

### Heatmap Analysis: Different genres have unique distributions of ratings. For example, genres like Children & Family Movies lean toward lower ratings (e.g., PG), while Dramas and Crime Shows often have higher ratings (e.g., TV-MA). This relationship provides insights into how genre and audience targeting influence content ratings.
